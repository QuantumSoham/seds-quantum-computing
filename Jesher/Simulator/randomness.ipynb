# Truly Random 🗿

```python
from qiskit import *
import ipywidgets as pyw
from IPython.display import display,clear_output
from qiskit.tools.visualization import plot_histogram
```

## A Truly Random bit 😂

```python
random_circuit = QuantumCircuit(1, 1)

#
random_circuit.h(0)

# 
random_circuit.measure(0, 0)

# 
backend = Aer.get_backend('qasm_simulator')# can use real backend if you want 
job = execute(random_circuit, backend, shots=1)
result = job.result()
bit = int(list(result.get_counts().keys())[0])

print("Random bit: ", bit)
random_circuit.draw('mpl')
```

## How about in the range of 0-100

```python
circuit = QuantumCircuit(7, 7)

#
for i in range(7):
    circuit.h(i)
# 
for i in range(7):
    circuit.measure(i,i)
# 
backend = Aer.get_backend('qasm_simulator')
job = execute(circuit, backend, shots=1)
result = job.result()
bit = int(list(result.get_counts().keys())[0],2)

while bit>100:
    job = execute(circuit, backend, shots=1)
    result = job.result()
    bit = int(list(result.get_counts().keys())[0],2)
print("Random Number: ", bit)

circuit.draw('mpl')
```
